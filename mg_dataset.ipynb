```python
import matplotlib.pyplot as plt
from os.path import exists, join
```

```python
ds_path = 'mg_data/ITER_like_equilibrium_dataset_sample.mat' # sample dataset
# ds_path = 'mg_data/ITER_like_equilibrium_dataset.mat' # full dataset
assert exists(ds_path), f"Dataset not found at {ds_path}"

#load teh mat files
import scipy.io as sio
data = sio.loadmat(ds_path)

print(data.keys())
# 'DB_meas_Bpickup_test_ConvNet', 'DB_psi_pixel_test_ConvNet', 'DB_res_RHS_pixel_test_ConvNet', 
# 'DB_Jpla_pixel_test_ConvNet', 'DB_coils_curr_test_ConvNet', 'RR_pixels', 'ZZ_pixels', 
# 'DB_separatrix_200_test_ConvNet', 'XP_YN', 'DB_f_test_ConvNet', 'DB_p_test_ConvNet']

idx = 1
lvs = 20

# measurements from B (magnetic field) pickup coils (n, 187)
print(f'DB_meas_Bpickup_test_ConvNet: {data["DB_meas_Bpickup_test_ConvNet"].shape}')
# psi (magnetic flux) pixel values (n, 64, 64)
print(f'DB_psi_pixel_test_ConvNet: {data["DB_psi_pixel_test_ConvNet"].shape}')   
plt.figure(); plt.imshow(data["DB_psi_pixel_test_ConvNet"][idx], cmap='plasma'); plt.colorbar(); plt.title('psi')
plt.contour(data["DB_psi_pixel_test_ConvNet"][idx], levels=lvs, cmap='plasma_r')
# residual right-hand side pixel values (n, 64, 64)          
print(f'DB_res_RHS_pixel_test_ConvNet: {data["DB_res_RHS_pixel_test_ConvNet"].shape}')     
plt.figure(); plt.imshow(data["DB_res_RHS_pixel_test_ConvNet"][idx], cmap='plasma'); plt.colorbar(); plt.title('res_RHS')
plt.contour(data["DB_res_RHS_pixel_test_ConvNet"][idx], levels=lvs, cmap='plasma_r')
# plasma current pixel values (n, 64, 64) 
print(f'DB_Jpla_pixel_test_ConvNet: {data["DB_Jpla_pixel_test_ConvNet"].shape}')  
plt.figure(); plt.imshow(data["DB_Jpla_pixel_test_ConvNet"][idx], cmap='plasma'); plt.colorbar(); plt.title('Jpla') 
plt.contour(data["DB_Jpla_pixel_test_ConvNet"][idx], levels=lvs, cmap='plasma_r')
# coil currents (n, 14)     
print(f'DB_coils_curr_test_ConvNet: {data["DB_coils_curr_test_ConvNet"].shape}')
plt.figure(); plt.bar(range(14), data["DB_coils_curr_test_ConvNet"][idx]); plt.title('coil currents')
# separatrix coordinates (n, 200, 2)   
print(f'DB_separatrix_200_test_ConvNet: {data["DB_separatrix_200_test_ConvNet"].shape}')
# R and Z coordinates (n, 64, 64)
print(f'RR_pixels: {data["RR_pixels"].shape}')
print(f'ZZ_pixels: {data["ZZ_pixels"].shape}')
#scatter the R and Z coordinates
plt.figure(); plt.scatter(data["RR_pixels"], data["ZZ_pixels"], s=3); plt.axis('equal'); plt.title('RZ')
# unclear
print(f'XP_YN: {data["XP_YN"].shape}')
print(f'DB_f_test_ConvNet: {data["DB_f_test_ConvNet"].shape}')
print(f'DB_p_test_ConvNet: {data["DB_p_test_ConvNet"].shape}')
n_samples = data["DB_meas_Bpickup_test_ConvNet"].shape[idx]
print(f'Number of samples: {n_samples}')
plt.show()
```
